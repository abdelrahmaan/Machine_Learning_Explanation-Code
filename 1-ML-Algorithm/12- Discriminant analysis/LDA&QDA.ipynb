# 1- Linear Discriminant Analysis

```python
#Applying LDA Model 

'''
sklearn.discriminant_analysis.LinearDiscriminantAnalysis(solver='svd,shrinkage=None,priors=None,
                                                         n_components=None,store_covariance=False,tol=0.0001)
'''
```

```python
#Import Libraries
from sklearn.datasets import load_breast_cancer

from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Example 1

```python
#load breast cancer data

BreastData = load_breast_cancer()

#X Data
X = BreastData.data

#y Data
y = BreastData.target
```

```python
#Splitting data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=44, shuffle =True)
```

```python
#Applying LDA Model 
LDAModel = LinearDiscriminantAnalysis(n_components=3)
LDAModel.fit(X_train, y_train)
```

```python
#Calculating Details
print('LDAModel Train Score is : ' , LDAModel.score(X_train, y_train))
print('LDAModel Test Score is : ' , LDAModel.score(X_test, y_test))
```

```python
#Calculating Prediction
y_pred = LDAModel.predict(X_test)
```

```python
#Calculating Confusion Matrix
CM = confusion_matrix(y_test, y_pred)
print('Confusion Matrix is : \n', CM)

# drawing confusion matrix
sns.heatmap(CM)
```

### Example 2

```python
dataset = pd.read_csv('iris.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Logistic Regression Classifier
from sklearn.linear_model import LogisticRegression
classifier1 = LogisticRegression(random_state = 0)
classifier1.fit(X_train, y_train)

y_pred1 = classifier1.predict(X_test)

score_train_1 = classifier1.score(X_train, y_train)
score_test_1  = classifier1.score(X_test, y_test)

print('score_train 1 = ', score_train_1 )
print('score_test 1  = ', score_test_1 )


from sklearn.metrics import confusion_matrix
cm1 = confusion_matrix(y_test, y_pred1)

print('cm1 \n' , cm1 )

import seaborn as sns
sns.heatmap(cm1, center=True)
plt.show()
```

```python
#Applying LDA Model 
LDAModel = LinearDiscriminantAnalysis(n_components=3)
LDAModel.fit(X_train, y_train)

#Calculating Details
print('LDAModel Train Score is : ' , LDAModel.score(X_train, y_train))
print('LDAModel Test Score is : ' , LDAModel.score(X_test, y_test))

#Calculating Prediction
y_pred = LDAModel.predict(X_test)

#Calculating Confusion Matrix
CM = confusion_matrix(y_test, y_pred)
print('Confusion Matrix is : \n', CM)

# drawing confusion matrix
sns.heatmap(CM)
```

### Example 1

### Example 3

```python
data = pd.read_csv('Smarket.csv', index_col=0, parse_dates=True)
data.head()
```

```python
X_train = data[:'2004'][['Lag1','Lag2']].values
y_train = data[:'2004']['Direction'].values

X_test = data['2005':][['Lag1','Lag2']].values
y_test = data['2005':]['Direction'].values

```

```python
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
pred=lda.predict(X_test)

print('Prediction for LDA : ',np.unique(pred, return_counts=True))
print('\nCM for LDA : \n',confusion_matrix(pred, y_test))
print('Report for LDA \n: ',classification_report(y_test, pred, digits=3))
```

```python
qda = QuadraticDiscriminantAnalysis()
qda.fit(X_train, y_train)
pred2 = qda.predict(X_test)

print('Prediction for QDA : ',np.unique(pred2, return_counts=True))
print('\nCM for QDA : \n',confusion_matrix(pred2, y_test))
print('Report for QDA : \n',classification_report(y_test, pred2, digits=2))
```
