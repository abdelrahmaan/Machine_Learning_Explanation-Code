## 1.1-  Mean Absolute Error

```python
from sklearn.metrics import mean_absolute_error
'''
MAE = mean_absolute_error(y_true, y_pred, multioutput='uniform_average')
multioutput = the defult value is 'uniform_average' or 'raw_values'  for each row

'''
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]

MAE = mean_absolute_error(y_true, y_pred)#default
print('Mean Absolute Error = ', MAE)
MAE = mean_absolute_error(y_true, y_pred , multioutput='uniform_average')
print('Mean Absolute Error = ', MAE)
print('-----------')
y_true = [
            [0.5, 1],
            [-1, 1],
            [7, -6]]
y_pred = [
            [0, 2], 
            [-1, 2], 
            [8, -5]]
MAE = mean_absolute_error(y_true, y_pred , multioutput='raw_values')
print('Mean Absolute Error = ', MAE)
```

## 1.2-  Mean Squared Error
* the commonly use

```python
from sklearn.metrics import mean_squared_error
'''
MAE = mean_squared_error(y_true, y_pred, multioutput='uniform_average')
multioutput = the defult value is 'uniform_average' or 'raw_values'  for each row

'''
y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]

MSE = mean_squared_error(y_true, y_pred)#default
print('Mean Squared Error = ', MSE)
MSE = mean_squared_error(y_true, y_pred , multioutput='uniform_average')
print('Mean Squared Error = ', MSE)
print('-----------')
y_true = [
            [0.5, 1],
            [-1, 1],
            [7, -6]]
y_pred = [
            [0, 2], 
            [-1, 2], 
            [8, -5]]
MSE = mean_squared_error(y_true, y_pred , multioutput='raw_values')
print('Mean Squared Error = ', MSE)
```

## 1.3- Mean Absolute Error

## 2.1- Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
y_pred = ['a','a','b','b','a','b','a','a','a','a']
y_true  = ['a','b','b','a','b','a','a','b','a','b']
CM = confusion_matrix(y_pred, y_true)
print(CM) # my actual values is the diagonal of the metrics
# drawing confusion matrix
sns.heatmap(CM, center= True)
plt.show()
```

## 2.2- Accuracy Score

```python
from sklearn.metrics import accuracy_score
#Accuracy Score:((TP + TN) / float(TP + TN + FP + FN))

y_pred = [0, 2, 1, 3,5,3]
y_true = [0, 1, 2, 3,5,3]

AS = accuracy_score(y_pred, y_true)
print('Accuracy Score is =' , AS)
print('-------------------------')
AS = accuracy_score(y_pred, y_true, normalize=False)# number of all trues
print('Accuracy Score is =' , AS)
```

## 2.3- Recall Score

```python
from sklearn.metrics import recall_score

# Recall Score : (Sensitivity) (TP / float(TP + FN))

y_pred =  ['a','b','c','a','b','c','a','b','c','a']
y_true =   ['a','a','b','b','a','b','c','c','b','b']

RecallScore = recall_score(y_pred, y_true, average='micro')
#it can be : binary,macro,weighted,samples

print('Recall Score is : ', RecallScore)
```

## 2.4- Precision Score

```python
from sklearn.metrics import precision_score

# Precision Score : (Specificity) #(TP / float(TP + FP))  

y_pred =  ['a','b','c','a','b','c','a','b','c','a']
y_true =   ['a','a','b','b','a','b','c','c','b','b']


PrecisionScore = precision_score(y_pred, y_true, average='micro') 
#it can be : binary,macro,weighted,samples

print('Precision Score all Score is : ', PrecisionScore)
```

## 2.5- F1 Score

```python
from sklearn.metrics import f1_score
# F1 Score  : 2 * (precision * recall) / (precision + recall)
# f1_score(y_true, y_pred, labels=None, pos_label=1, average=binary, sample_weight=None)

y_pred = [0, 2, 1, 0, 0, 1]
y_true = [0, 1, 2, 0, 1, 2]

F1Score = f1_score(y_true, y_pred, average='micro')
#it can be : binary,macro,weighted,samples

print('F1 Score is : ', F1Score)
```

## 2.8- ROC Curve 'Receiver operating characteristic'
* use for binary classification

```python
from sklearn.metrics import roc_curve
# Receiver Operating Characteristic :  
#roc_curve(y_true, y_score, pos_label=None, sample_weight=None,drop_intermediate=True)

y =      np.array([1    , 1     , 2     , 2])
scores = np.array([0.1  , 0.4   ,   0.35, 0.8])
fpr, tpr, thresholds = roc_curve(y, scores, pos_label=2)
# fpr, tpr ==> false/true positive rate
print('fpr Value  : ', fpr)
print('tpr Value  : ', tpr)
print('thresholds Value  : ', thresholds)
```

 ## 2.9- AUC  'Area Under The Curve'


```python
from sklearn.metrics import roc_curve , auc

y =      np.array([1    , 1     , 2     , 2])
scores = np.array([0.1  , 0.4   ,   0.35, 0.8])

fpr, tpr, thresholds = roc_curve(y, scores, pos_label=2)
# fpr, tpr ==> false/true positive rate
AUCValue = auc(fpr, tpr)
print('AUC Value  : ', AUCValue) # An excellent model has AUC near to the 1
```

* `ROC` is a probability curve and `AUC` represents degree or measure of separability. 
* `It tells how much model is capable of distinguishing between classes.` 
* An excellent model has `AUC near to the 1`
