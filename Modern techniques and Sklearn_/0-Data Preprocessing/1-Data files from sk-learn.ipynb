## Iris Data

```python
from sklearn.datasets import load_iris
iris_data = load_iris()
# X data
X = iris_data.data
print('X data is\n{}\n'.format(X[:10]))
print('X shape is\n{}\n\n'.format(X.shape))
print('X Feature are\n{}\n'.format(iris_data.feature_names))
print('------------------------\n')
 
# y data
y = iris_data.target
print('y data is \n{}\n'.format(y[40:60]))
print('y shape is \n{}\n'.format(y.shape))
print('y Columms are\n{}\n'.format(iris_data.target_names))
```

## Data for  Regression

```python
from sklearn.datasets import make_regression
# load regression data
X, y = make_regression(n_samples= 1000, n_features= 10 , shuffle= True)

print('X data is\n{}\n'.format(X[:5]))
print('X shape is\n{}\n\n'.format(X.shape))
print('------------------------\n')
 
print('y data is \n{}\n'.format(y[:10]))
print('y shape is \n{}\n'.format(y.shape))
```

## Data for  Classification

```python
from sklearn.datasets import make_classification
# load classification data
X, y = make_classification(n_samples= 100, n_features= 5 , shuffle= True)

print('X data is\n{}\n'.format(X[:5]))
print('X shape is\n{}\n\n'.format(X.shape))
print('------------------------\n')
 
print('y data is \n{}\n'.format(y[:10]))
print('y shape is \n{}\n'.format(y.shape))
```
